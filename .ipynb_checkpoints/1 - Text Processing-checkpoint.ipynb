### Soal 3 : analisa untuk fitur “content”
<p>a.Hitunglah jumlah kata yang unik dalam korpus. Catatan: Kata “Jakarta” dan “jakarta” dianggap sebagai kata yang sama</p>

```python
import pandas as pd
import nltk
nltk.download("punkt")
```

```python
df = pd.read_json("Korpus.json")
#Membuat dictionary list of sentence from content
sentence_dict={}
x = 0
for berita in df.loc[:,"content"]:
    sent_text = nltk.sent_tokenize(berita)
    sentence_dict[x] = sent_text
    x+=1
#sentence_dict
```

```python
sentence_list = []

for i in sentence_dict:
    for sentence in sentence_dict[i]:
        sentence_list.append(sentence.lower())
        
# sentence_list
```

```python
#Tokenisasi kata dari semua kalimat pada sentence_list. lalu dimasukkan ke word_list
word_list = []

for sentence in sentence_list:
    tokenized_word = nltk.word_tokenize(sentence)
    for i in tokenized_word:
        if i.isalnum():
            word_list.append(i)
```

```python
#Menjadikan word_list sebagai dataframe word_df dan hitung value uniknya
word_df = pd.DataFrame(word_list, columns=["word"])
word_df_2 = word_df["word"].value_counts().to_frame(name="count")

word_df_2
print("Jumlah kata yang unik adalah : " + str(len(word_df_2)))
```

# Bagian3 : TOKENIZATION

```python
f = open("Korpus_Ringkas.json", "r")
data2 = json.loads(f.read())
# data2
```

1. Transformasi semua karakter dalam “Korpus_Ringkas.json” ke format
lowercase!

```python
data_list = []
for obj in data2:
    if obj["content"] not in data_list:
        data_list.append(obj["content"].lower())   
```

2. Hilangkan seluruh karakter berupa angka, whitespace berlebih, dan tanda baca!

```python
result_list=[] 
for i in data_list:
    r = re.findall('[^0-9\.\,\(\)\/\:\?\!\;\-\"]',i)
    result = ''.join(r)
    result_list.append(result)
```


3.Lakukan tokenisasi pada setiap baris data. Tampilkan hasil tokenisasi berupa
list of tokens dari 5 data pertama!

```python
content_list=[]
for i in result_list:
    sent_text = nltk.sent_tokenize(i)
    content_list.append(sent_text)
    
content_list


final_result = []

for i in content_list:
    for sentence in i:
        for word in i:
            final_result.append(word)
```

4. Untuk setiap baris data tentukan token yang dapat dikategorikan sebagai stopwords! Hilangkan kata-kata tersebut dari list of tokens!

```python
 from nltk.corpus import stopwords

cachedStopWords = stopwords.words("indonesian")
# print(cachedStopWords)

# pattern = re.compile(r'\b(' + r'|'.join(cachedStopWords) + r')\b\s*')
# text = pattern.sub('', final_result)
# text

exclude_stopword =[]
for sentence in final_result:
    sent_word = nltk.word_tokenize(sentence)
    for word in sent_word:
        if word not in cachedStopWords:
            exclude_stopword.append(word)
    
# exclude_stopword
```

1. Menggunakan list of tokens yang diperoleh dari bagian C, lakukan stemming dengan menggunakan library Sastrawi untuk masing-masing token pada setiap baris data.
```python
# import StemmerFactory class
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
 
#  create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
 

# stemming process
result_stemming = []

for i in final_result:
    output = stemmer.stem(i)
    result_stemming.append(output)
```
